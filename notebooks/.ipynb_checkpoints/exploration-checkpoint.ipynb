# Exploration of Complaint Clusters

This notebook explores the clustered consumer complaint dataset produced by the NLP pipeline.  
The goal is to understand the themes represented by each cluster through:

- visualizing cluster distribution  
- inspecting sample complaints  
- extracting common keywords  
- summarizing cluster themes  

This analysis supports the development-phase explanation and the final report.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/consumer_complaints_with_clusters.csv")
df.head()
```

## Dataset Overview

We begin by inspecting the structure of the clustered dataset to confirm that the preprocessing and clustering pipeline executed correctly.


```python
df.info()
```

```python
df.describe(include="all")
```

## Distribution of Complaints Across Clusters

This visualization shows how many complaints fall into each cluster.  
It helps identify dominant themes and ensures that clustering is balanced.


```python
plt.figure(figsize=(8,4))
df["cluster"].value_counts().sort_index().plot(kind="bar", color="steelblue")
plt.title("Number of Complaints per Cluster")
plt.xlabel("Cluster")
plt.ylabel("Count")
plt.show()
```

## Sample Complaints per Cluster

To interpret each cluster, we examine a few representative complaints.  
This provides qualitative insight into the themes captured by the model.


```python
for c in sorted(df["cluster"].unique()):
    print(f"\n=== Cluster {c} ===")
    print(df[df.cluster == c]["Consumer complaint narrative"].head(3).to_string(index=False))
```

## Top Words per Cluster

To further understand each cluster, we extract the most frequent words from the cleaned text.  
This helps identify recurring concepts and supports theme interpretation.


```python
df["clean_text"] = df["clean_text"].fillna("").astype(str)
```

```python
from collections import Counter
import re

# Ensure all text is string
df["clean_text"] = df["clean_text"].fillna("").astype(str)

def get_top_words(texts, n=10):
    words = " ".join(texts).lower()
    words = re.findall(r"[a-zA-Z]+", words)
    return Counter(words).most_common(n)

for c in sorted(df["cluster"].unique()):
    texts = df[df.cluster == c]["clean_text"].tolist()
    print(f"\n=== Cluster {c} — Top Words ===")
    print(get_top_words(texts, n=12))
```

## Cluster Summaries

To support the interpretation of each cluster, we generate concise summaries based on the most frequent words appearing in the cleaned text of each cluster.  
These summaries help identify the dominant themes and will be used in the development-phase explanation and final report.

The process:
- ensure all cleaned text is valid (no missing values)
- extract the top 15 most common words per cluster
- combine these into a short keyword-based summary


```python
# Ensure all cleaned text is string and NaN-free
df["clean_text"] = df["clean_text"].fillna("").astype(str)

cluster_summaries = {}

for c in sorted(df["cluster"].unique()):
    texts = df[df.cluster == c]["clean_text"].tolist()
    words = [w for w, _ in get_top_words(texts, n=15)]
    cluster_summaries[c] = ", ".join(words)

cluster_summaries
```
